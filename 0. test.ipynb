```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


range = (- 3 * np.pi, 3 * np.pi)

x = np.linspace(range[0], range[1], 1000)
g = pd.Series(np.sin(x), index=x)
g2 = pd.Series(np.cos(x), index=x)
g.plot()
g2.plot(title="Sin Cos", grid=True)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.xticks(np.arange(-3 * np.pi, 3.1 * np.pi, np.pi), [r'$-3\pi$', r'$-2\pi$', r'$-\pi$', '0', r'$\pi$', r'$2\pi$', r'$3\pi$'])
plt.show()
```

```python
"""
- 단순 선형 회귀 예시
    |hours(x)|score(y)|
    |--------|--------|
    |2       |81      |
    |4       |93      |
    |6       |91      |

"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


d = pd.Series([25, 50, 42, 61], index=[2,3,4,5])

# h1 = pd.Series([0, 70], index=[2, 4])
# h2 = pd.Series([8, 70], index=[2, 5])
# h3 = pd.Series([40, 0], index=[2, 4])


# h1.plot(color='red')
# h2.plot(color='black')
# h3.plot(color='green')


ca = {
    'w': None,
    'b': None,
    'mse': None
}
for w in np.linspace(-99, 99):
    for b in np.linspace(-99, 99):
        mse = ((d - (w * d.index + b))**2).mean()
        if(ca['mse'] == None or ca['mse'] > mse):
            ca['mse'] = mse
            ca['w'] = w
            ca['b'] = b

x = np.linspace(2, 5)

pd.Series(ca['w'] * x + ca['b'], index=x).plot()

d.plot(style='bo', grid=True, title=rf'$y = {ca['w']:.4f}x + {ca['b']:.4f}$')

print(d.mean())

plt.xlabel('hours')
plt.ylabel('score')

for x_val, y_val in d.items():
    plt.plot([x_val, x_val], [y_val, ca['w']*x_val + ca['b']], 'r--')
```

```python
"""
- 단순 선형 회귀 예시
    |hours(x)|score(y)|
    |--------|--------|
    |2       |81      |
    |4       |93      |
    |6       |91      |

"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


d = pd.Series([25, 50, 42, 61], index=[2,3,4,5])

ca = {
    'w': [],
    'mse': []
}
for w in np.linspace(-50, 75):
    ca['mse'].append(((d - (w * d.index))**2).mean())
    ca['w'].append(w)

pd.Series(ca['mse'], index=ca['w']).plot()

plt.xlabel(f'$w$')
plt.ylabel(r'$cost$')

```
