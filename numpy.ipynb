### 배열의 연산
- 기본적으로 element-wise 연산임. 동일 위치의 요소끼리 + - * / 계산
- 브로드캐스팅 가능. 서로 다른 크기의 배열 간 연산이 가능함. 단, 다음 조건을 만족할 때
  - 각 배열의 오른쪽 차원부터 같거나 둘 중 하나가 1이어야 함.

- $\begin{bmatrix} 1 \\ 2 \end{bmatrix} + \begin{bmatrix} 1 & 2 & 3 & 4 \end{bmatrix}$
$= \begin{bmatrix} 1 & 1 & 1 & 1 \\ 2 & 2 & 2 & 2 \end{bmatrix} + \begin{bmatrix} 1 & 2 & 3 & 4 \\ 1 & 2 & 3 & 4 \end{bmatrix}$
$= \begin{bmatrix} 1+1 & 1+2 & 1+3 & 1+4 \\ 2+1 & 2+2 & 2+3 & 2+4 \end{bmatrix}$
$= \begin{bmatrix} 2 & 3 & 4 & 5 \\ 3 & 4 & 5 & 6 \end{bmatrix}$  
- 차원: (2, 1) + (4,) -> 4 랑 1 비교 => 브로드캐스팅 가능

- 브로드캐스팅이 안 되는 경우 ValueError 발생



### Array 슬라이싱 (Slicing)
NumPy 배열의 슬라이싱은 Python 리스트와 유사하지만, 다차원에서 더 강력합니다. 슬라이싱은 원본 배열의 뷰(view)를 반환하므로 메모리를 효율적으로 사용.

- **기본 1D 슬라이싱**: `arr[start:stop:step]`
  - `start`: 시작 인덱스 (포함)
  - `stop`: 끝 인덱스 (제외)
  - `step`: 단계 (기본 1)
  - 예: `arr[1:5:2]` → 인덱스 1, 3 선택

- **다차원 슬라이싱**: `arr[row_slice, col_slice, ...]`
  - 각 차원별로 슬라이스 적용
  - 예: `arr[1:3, 0:2]` → 행 1~2, 열 0~1 선택

- **특별한 인덱스**:
  - `:` : 전체 범위
  - `::2` : 2칸씩
  - `::-1` : 역순
  - `...` : 생략된 차원

- **뷰 vs 복사**: 슬라이싱은 뷰를 반환하므로 원본 변경 시 영향을 받음. 복사하려면 `.copy()` 사용.

### 마스킹 (Masking)
마스킹은 조건에 따라 배열 요소를 선택. 주로 불리언 배열을 사용.

- **불리언 마스킹**: `arr[condition]`
  - `condition`: True/False 배열 또는 조건식. 배열의 관계/논리 연산자로 생성 가능.
  - 예: `arr[arr > 5]` → 5보다 큰 요소만 선택
  - 다차원에서도 작동: `arr[arr[:, 0] > 2]` → 첫 번째 열이 2보다 큰 행 선택

- **np.where**: 조건에 따라 값 선택
  - `np.where(condition, x, y)`: 조건이 True면 x, False면 y
  - `np.where(condition)`: 조건이 True인 인덱스 튜플 반환
  - 예: `np.where(arr > 5, arr, 0)` → 5보다 크면 그대로, 아니면 0

- **고급 마스킹**:
  - 복합 조건: `arr[(arr > 2) & (arr < 8)]`
  - 인덱스 마스킹: `arr[[1, 3, 5]]` → 특정 인덱스 선택
  - 팬시 인덱싱: 다차원 인덱스 사용

마스킹은 데이터 필터링, 조건부 연산 등에 사용
