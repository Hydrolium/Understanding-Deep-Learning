## 웹 크롤링
- 주로 웹 페이지의 링크 구조를 탐색하고 웹사이트의 다양한 페이지를 방문하며 데이터를 포괄적으로 수집 및 저장하는 작업
- ex. 특정 쇼핑몰의 어러 제품 정보를 수집 / 뉴스 사이트의 홈페이지부터 시작해 최신 뉴스 기사 수집 후 데이터 베이스에 저장하는 작업

## 웹 스크래핑
- 특정 웹 페이지에서 필요한 정보를 추출하는 작업
- ex. 특정 쇼핑몰의 제품 페이지에서 제품명, 가격, 리뷰 수 등 필요한 정보만 추출하는 작업

## HTML 파싱
- 웹페이지의 HTML을 분석하여 필요한 URL을 추출하는 작업
- 파싱(parsing): 특정 텍스트를 분석하여 그 데이터로부터 필요한 정보를 추출하는 과정

```python
import urllib.request

url = 'https://www.python.org/static/img/python-logo.png'
file_name = 'data/saved/python-logo.png'

urllib.request.urlretrieve(url, file_name)
print('download ok!!')
```

```python
import pandas as pd

url = 'data/sample.html'
tables = pd.read_html(url)

for i in range(len(tables)):
    print(f"tables[{i}]")
    print(tables[i])
    print()
    print()

df = tables[1]
df.set_index(['name'], inplace=True)

print(df)
```

```python
import urllib.request
import re

url = "http://finance.naver.com/item/main.nhn?code=005930"

html = urllib.request.urlopen(url)

html_contents = str(html.read().decode("utf-8"))

stock_results = re.findall(r'(<dl class="blind">)([\s\S]+?)(</dl>)', html_contents)

samsung_stock = stock_results[0]

index_list = re.findall(r'(<dd>)([\s\S]+?)(</dd>)', samsung_stock[1])

for index in index_list:
    print(index[1])
```
