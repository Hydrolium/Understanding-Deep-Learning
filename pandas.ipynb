# Pandas
- 데이터 분석을 위한 파이썬 라이브러리. 대용량 데이터 처리에 강력
- Series(1차원), DataFrame(2차원) 구조 제공
- Series는 np.array와 유사하지만, 인덱스와 데이터가 함께 관리됨(딕셔너리와 닮음.)

### Pandas 주요 함수 및 메서드

**Series 생성 및 속성**
| 함수/메서드 | 설명 | 반환형 |
|------|------|------|
| `pd.Series(data, index=None, name=None)` | Series 생성 (data: 리스트, 딕셔너리, 스칼라 등), data가 스칼라면 인덱스의 모든 값을 data로 설정함. | Series |
| `series.values` | Series의 데이터 배열 반환 | numpy.ndarray |
| `series.index` | Series의 인덱스 반환 | Index |
| `series.dtype` | Series의 데이터 타입 | dtype |
| `series.name` | Series의 이름 (읽기/쓰기) | str or None |

**DataFrame 생성 및 기본 조작**
| 함수/메서드 | 설명 | 반환형 |
|------|------|------|
| `pd.DataFrame(data, index=None, columns=None)` | DataFrame 생성 | DataFrame |
| `df[column_name]` | 열 선택 (Series 반환) | Series |
| `df[[col1, col2]]` | 여러 열 선택 (DataFrame 반환) | DataFrame |
| `df.column_name` | 열 접근 (속성 방식) | Series |
| `df.index` | 행 인덱스 | Index |
| `df.columns` | 열 이름 | Index |
| `df.dtypes` | 각 열의 데이터 타입 | Series |
| `df.describe()` | 수치형 데이터의 통계 정보 | DataFrame |

**인덱싱 및 선택**
| 함수/메서드 | 설명 | 반환형 |
|------|------|------|
| `df.loc[label]` | 라벨(인덱스명)로 행/열 선택 (끝 포함) | Series or DataFrame |
| `df.loc[start:end, col]` | 라벨 슬라이싱 (끝 포함) | Series or DataFrame |
| `df.iloc[position]` | 정수 위치로 행/열 선택 (끝 미포함) | Series or DataFrame |
| `df.iloc[start:end, start:end]` | 정수 위치 슬라이싱 (끝 미포함) | DataFrame |
| `df[df[col] >= value]` | 불리언 마스킹으로 행 필터링 | DataFrame |

**행/열 추가 및 삭제**
| 함수/메서드 | 설명 | 반환형 |
|------|------|------|
| `df[new_col] = value` | 새 열 추가 (스칼라, 배열, Series 모두 가능) | None (제자리 수정) |
| `df.loc[new_label, :] = values` | 라벨로 새 행 추가 | None (제자리 수정) |
| `df.drop(label, axis=0, inplace=False)` | 행(axis=0) 또는 열(axis=1) 삭제 (copy 반환) | DataFrame or None |
| `del df[column]` | 열 즉시 삭제 | None |

**인덱스 및 재구성**
| 함수/메서드 | 설명 | 반환형 |
|------|------|------|
| `pd.date_range(start, periods=None, freq='D')` | 날짜 범위 생성 | DatetimeIndex |
| `pd.to_datetime(arg, format=None)` | 문자열/숫자를 datetime 객체로 변환 | Timestamp or Series or DatetimeIndex |
| `df.index = new_index` | 인덱스 변경 | None (제자리 수정) |
| `df.index.name = "name"` | 인덱스 이름 지정 | None (제자리 수정) |
| `df.columns.name = "name"` | 열 이름(컬럼명의 이름) 지정 | None (제자리 수정) |
| `df.reindex(index=..., columns=...)` | 인덱스와 열 순서 재구성 (새 DataFrame 반환) | DataFrame |
| `pd.permutation(x)` | 배열/범위를 무작위 순서로 섞음 | numpy.ndarray |

**정렬**
| 함수/메서드 | 설명 | 반환형 |
|------|------|------|
| `df.sort_index(axis=0, ascending=True)` | 인덱스로 정렬 (axis=0: 행, axis=1: 열) | DataFrame or None |
| `df.sort_values(by=col_name, ascending=True)` | 특정 열 값으로 정렬 | DataFrame or None |
| `df.sort_values(by=[col1, col2], ascending=[True, False])` | 다중 열 정렬 (우선순위, 각각 방향 지정 가능) | DataFrame or None |

**통계 및 집계**
| 함수/메서드 | 설명 | 반환형 |
|------|------|------|
| `df.count()` | 각 열의 비-NaN 데이터 개수 | Series |
| `df.sum(axis=None)` | 합계 (axis=None: 전체, axis=0: 열단위, axis=1: 행단위) | Series or scalar |
| `df.mean(axis=None)` | 평균 | Series or scalar |
| `df.median(axis=None)` | 중앙값 | Series or scalar |
| `df.std(axis=None)` | 표준편차 | Series or scalar |
| `df.var(axis=None)` | 분산 | Series or scalar |
| `df.min(), df.max()` | 최소, 최대값 | Series or scalar |
| `df.idxmin(), df.idxmax()` | 최소/최대값의 인덱스 반환 | Series |
| `df.quantile(q=0.5, axis=None)` | 사분위수 (q=0.25, 0.5, 0.75 등) | Series or DataFrame |
| `df.cumsum(axis=None)` | 누적합 | DataFrame or Series |
| `df.cumprod(axis=None)` | 누적곱 | DataFrame or Series |
| `df.abs()` | 절댓값 | DataFrame |
| `df.corr()` | 상관계수 행렬 | DataFrame |
| `series1.corr(series2)` | 두 Series 간 상관계수 | float |
| `series1.cov(series2)` | 두 Series 간 공분산 | float |

**Key Points**
- `loc[]`: 라벨 기반, 끝 포함 (inclusive)
- `iloc[]`: 정수 위치 기반, 끝 미포함 (exclusive)
- `drop()`: 새로운 복사본 반환 (view 아님), `inplace=True`로 원본 수정 가능
- DataFrame에 Series 추가: 인덱스 기반으로 정렬됨
- 불리언 조건 필터링: `df[column] operator value` 형태

```python
df14 = pd.DataFrame({
    'num': [1,2,3,4],
    'year': [2021, 2022, 2026, 2025],
    'point': [1.0, 3.0, 2.0, 5.0]
}, index=pd.date_range('20260325', periods=4))

suf = np.random.permutation(df14.index)
df15 = df14.reindex(index=suf, columns=np.random.permutation(df14.columns)) # 새롭게 인덱스와 열을 바꾼 DF 반환
df15
```
