```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import zipfile
```

```python

G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2, 3, 4, 5])
G.add_edge(2, 1)
G.add_edges_from([(3, 1),(4, 1),(5, 1)])
G.add_edges_from([(4, 5),(5, 2),(2, 3),{3,4}])

G.remove_edge(2,3)

G.remove_node(2)
G.remove_nodes_from([2,3])
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='grey')
```

```python
G = nx.karate_club_graph()
degree = [d for n, d in G.degree()]
nx.draw(G, node_size=[n*100 for n in degree],with_labels=True, node_color='lightblue', edge_color='yellow')
```

```python
zf = zipfile.ZipFile('data/football.zip')
txt = zf.read('football.txt').decode()
gml = zf.read('football.gml').decode() # grapical markup language
gml = gml.split('\n')[1:]
G = nx.parse_gml(gml)

nx.draw(G, with_labels=True, node_color='lightblue', edge_color='yellow')
```

```python
K = np.loadtxt('data/test.csv', delimiter=',')
G = np.to_networkx_graph(K)
A = (G.subgraph(c) for c in nx.connected_components(G))
"""
nx.connected_components(G) : 연결된 노드들끼리의 리스트 반환
G.subgraph(c) : subgraph 반환
"""
A = list(A)[0]
nx.draw(A, with_labels=True, node_color='lightblue', edge_color='yellow')
```
