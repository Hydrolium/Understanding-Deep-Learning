```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

idxs = 'abcdefghijklmnopqrstuvwxyz'
k = pd.Series(np.random.randint(0, 100, len(idxs)), index=list(idxs))
# keywords = {'이것': 5, '단어': 5, '빈도수': 3}

wordcloud = WordCloud(font_path='data/NanumGothic.ttf',
                      background_color='white')

wordcloud = wordcloud.generate_from_frequencies(k)

fig = plt.figure()
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
```
