### JSON
- JSON(JavaScript Object Notation)은 XML보다 데이터 용량이 적고 코드로의 전환이 쉬워 XML의 대체재로 가장 많이 활용되고 있음.(2010 년 이후로 XML보다 많이 사용됨)
- 키-값의 쌍으로 구성됨(딕셔너리와 유사)

```python
import json

with open('data/json_example.json', 'r', encoding='utf8') as f:
    content = f.read()
    json_data = json.loads(content)

print(json_data)
```

```python
import json

dict_data = {'a': 1, 'b': [1,2,3,4]}

with open('data/saved/data.json', 'w') as f:
    json.dump(dict_data, f)
```
