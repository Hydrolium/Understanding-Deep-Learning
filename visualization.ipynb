```python
exam_data = {
    '이름': ['A', 'B', 'C'],
    '수학': [1,2,3],
    '영어': [1,2,3]
}
# , inplace=True
df = pd.DataFrame(exam_data)
ndf = df.set_index(['수학', '영어']) # 멀티인덱스
ndf = df.set_index(['수학']) # 멀티인덱스

# 정규화: 최소 0 최대 1
"""
(a-min)/(max-min), (b-min)/(max-min), (c-min)/(max-min)
면 최소가 0, 최대가 1이 된다.
"""
ndf['영어'] = ndf['영어'].astype('float')
minx = ndf['영어'] - ndf['영어'].min()
minmax = ndf['영어'].max() - ndf['영어'].min()
ndf['영어'] = minx/ minmax
ndf.describe()
```
