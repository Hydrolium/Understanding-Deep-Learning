```python
import folium

# location=[위도, 경도], tiles=지도모양, zoom_start=배율
# 일반적인 지도 형태
seoul_map1 = folium.Map(location=[37.55, 126.98],
                       tiles='OpenStreetMap',
                       zoom_start=12)

# 도로를 확실하게
seoul_map2 = folium.Map(location=[37.55, 126.98],
                       tiles='Cartodb Positron',
                       zoom_start=12)

# 다크모드
seoul_map3 = folium.Map(location=[37.55, 126.98],
                       tiles='Cartodb dark_matter',
                       zoom_start=12)
seoul_map1.save('data/saved/seoul1.html')
seoul_map2.save('data/saved/seoul2.html')
seoul_map3.save('data/saved/seoul3.html')
```

```python
import folium
import pandas as pd

df = pd.read_excel("data/서울지역 대학교 위치.xlsx", index_col=0)

seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)

for name, lat, lng in zip(df.index, df.위도, df.경도):
    # folium.Marker([lat, lng], popup=name).add_to(seoul_map)
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        color='brown',
        fill=True,
        fill_color='coral',
        fill_opacity=0.7,
        popup=name).add_to(seoul_map)

seoul_map.save('data/saved/seoul_universities.html')
```

```python
import folium
import pandas as pd
import json

df = pd.read_excel("data/경기도인구데이터.xlsx", index_col=0)
df.columns = df.columns.map(str)

geo_path = 'data/경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))

g_map = folium.Map(location=[37.55, 126.98], zoom_start=9)

year = '2017'

folium.Choropleth(geo_data,
                  data=df[year],
                  columns=[df.index, df[year]],
                  fill_color='YlOrRd', #노랑 -> 빨강 그라데이션
                  fill_opacity=0.7,
                  line_opacity=0.3,
                  threshold_scale=[10000, 100000, 300000, 500000, 700000],
                  key_on='feature.properties.name',
                  legend_name="인구 수").add_to(g_map)

g_map.save('data/saved/경기도인구'+year+'.html')
```
