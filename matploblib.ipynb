# Matplotlib
- 다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 파이썬 라이브러리.
- matplotlib.pyplot 모듈이 가장 많이 사용됨. MATLAB과 유사한 인터페이스 제공
- numpy, pandas의 자료구조를 쉽게 시각화 가능

## 주요 함수/메서드

| 함수/메서드 | 설명 | 반환값 |
|-----------|------|-------|
| **Figure 및 Subplot 생성** | | |
| plt.figure() | 새로운 Figure 객체 생성. 크기, dpi 등 설정 가능 | Figure |
| plt.subplots(nrows, ncols) | 지정된 행/열 수만큼 Subplot을 가진 Figure와 Axes 배열 생성. 한번에 여러 그래프 생성 | Figure, ndarray(Axes) |
| fig.add_subplot(nrows, ncols, index) | Figure 내에 새로운 Subplot 추가. index는 1부터 시작 | Axes |
| **기본 플롯 함수** | | |
| plt.plot(x, y) | 선 그래프 그리기. 현재 활성 Axes에 그림 추가 | Line2D |
| ax.plot(x, y) | Axes 객체에 선 그래프 그리기 | Line2D |
| plt.scatter(x, y) | 산점도 그리기 | PathCollection |
| ax.scatter(x, y) | Axes에 산점도 그리기 | PathCollection |
| Series.plot(kind='line') | Pandas Series/DataFrame 플롯. kind로 그래프 유형 선택 (line/bar/barh/hist 등) | Axes |
| s.plot(kind='bar') | 막대 그래프 (세로) | Axes |
| s.plot(kind='barh') | 막대 그래프 (가로) | Axes |
| Series.hist(bins=10) | 히스토그램 그리기. bins로 구간 개수 설정 | Axes |
| **축(Axes) 설정** | | |
| ax.set_xlim(xmin, xmax) | x축 범위 설정 | None |
| ax.set_ylim(ymin, ymax) | y축 범위 설정 | None |
| ax.get_xlim() | 현재 x축 범위 반환 | tuple(xmin, xmax) |
| ax.get_ylim() | 현재 y축 범위 반환 | tuple(ymin, ymax) |
| ax.set_xticks(ticks) | x축 주요 눈금 위치 설정 | None |
| ax.set_yticks(ticks) | y축 주요 눈금 위치 설정 | None |
| ax.set_xticklabels(labels) | x축 눈금 라벨 설정. rotation(회전), fontsize, color 등 옵션 가능 | None |
| ax.set_yticklabels(labels) | y축 눈금 라벨 설정 | None |
| **제목 및 라벨** | | |
| ax.set_title(title) | 그래프 제목 설정 | Text |
| ax.set_xlabel(label) | x축 레이블 설정 | Text |
| ax.set_ylabel(label) | y축 레이블 설정 | Text |
| **범례** | | |
| ax.legend(loc='best') | 범례 표시. loc로 위치 선택 (best/upper left/upper right/lower left/lower right 등). Series/DataFrame의 label이 필요 | Legend |
| **플롯 스타일 옵션** | | |
| color 또는 c | 색상 지정: b(blue), g(green), r(red), c(cyan), m(magenta), y(yellow), k(black), w(white) | - |
| marker | 마커 스타일: .(점), ,(픽셀), o(원), v(삼각형아래), ^(삼각형위), <(삼각형좌), >(삼각형우), 8(팔각형), s(정사각형), p(오각형), *(별), h(육각형), +(십자), x(x), D(대각마름모) | - |
| linestyle 또는 ls | 선 스타일: -(실선), --(점선), -.(점-실선), :(점), None(선 없음) | - |
| alpha | 투명도: 0(투명)~1(불투명). 기본값은 1 | - |
| stacked | 누적 막대 그래프: True로 설정 시 막대들이 누적됨 | - |

```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(axis=0), columns=['a', 'b', 'c', 'd'], index=np.arange(0, 100, 10))
df.plot()
```

```python
s2 = pd.Series(np.random.rand(16), index=list("abcdefghijklmnop"))
s2.plot(kind='bar')

s2.plot(kind='barh')
```

```python
df2 = pd.DataFrame(np.random.rand(6,4),
                   index=['one','two','three','four','five','six'],
                   columns=pd.Index(['A','B','C', 'D'], name="Genus"))

df2.plot(kind='bar')
```

```python
df21 = pd.DataFrame(np.random.rand(6,4),
                   index=['one','two','three','four','five','six'],
                   columns=pd.Index(['A','B','C', 'D'], name="Genus"))

df21.plot(kind='barh', stacked=True)
```

```python
s3 = pd.Series(np.random.normal(0, 1, size=200))
s3.hist()
```

```python
s31 = pd.Series(np.random.normal(0, 1, size=200))
s31.hist(bins=50) # 구간 갯수 설정. 기본값은 10
```

```python
x1 = np.random.normal(1, 1, size= (100, 1))
x2 = np.random.normal(-2, 4, size=(100, 1))
X = np.concatenate((x1, x2), axis=1)

df3 = pd.DataFrame(X, columns=["x1", 'x2'])

plt.scatter(df3['x1'], df3['x2'])

df3.x1.corr(df3.x2)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 2)
ax2 = fig.add_subplot(2, 2, 3)
ax3 = fig.add_subplot(2, 2, 4)
ax4 = fig.add_subplot(2, 2, 1)
# figure 를 2x2로 나누며, 좌측상단에서 우측방향으로 붙은 숫자 중 1번을 선택한다
# 1 2
# 3 4


plt.plot(np.random.randn(50).cumsum()) # 위치 지정 안하면 맨 마지막 추가한 subplot에 그림이 추가됨.
plt.plot(np.random.randn(50).cumsum()) # 덮어 쓰기가 아니라 계속 추가됨.

ax1.plot(np.random.randn(50).cumsum())
ax2.scatter(np.arange(30), np.arange(30) * 3*np.random.randn(30))

pd.Series([1,2,3,4,5]).plot(kind="bar", ax=ax3)
```

```python
fig1, axes = plt.subplots(2, 3)

for i in range(2):
    for j in range(3):
        axes[i][j].plot(np.random.randn(50).cumsum(), color='g', marker='o', linestyle='None')

plt.plot(np.random.randn(50).cumsum(), 'k.-', alpha=0.2) # 한번에 지정 가능, 투명도 설정 가능

# color: b(blue), g(green), r(red), c(cyan), m(magenta), y(yellow), k(black), w(white)
# marker: .(point), ,(pixel), o(circle), v(triangle_down), ^(triangle_up), <(triangle_left), >(triangle_right), 8(octagon), s(square), p(pentagon), *(star), h(hexagon), +(plus), x(x), D(diamond)
# linestyle: -(solid line), --(dashed line), -.(dash-dotted line), :(dotted line), None(draw nothing)
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

# 눈금 범위 지정
ax.set_xticks([0,250,500,750,1000])
ax.set_yticks([-40, -20, 20, 40])
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), label='one')
ax.plot(np.random.randn(1000).cumsum(), label='two')
ax.plot(np.random.randn(1000).cumsum(), label='three')

# 눈금 라벨 지정
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=0, fontsize='small', color='orange')
ax.set_title("TITLE")
ax.set_xlabel("X")
ax.set_ylabel("Y")

ax.legend(loc='best') # 범례 추가, 각 그래프에 label이 있어야 함.

ax.get_xlim()
ax.set_xlim([-100, 1200])
ax.set_ylim([-100, 100])
```
